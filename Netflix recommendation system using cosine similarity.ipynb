```python
# For finding the cosine similarity we need to merge the important columns.
# Type, Title, Director, Cast, Description, Country are the column considered for Cosine Similarity
# Why bacause while recommendation, machine learning model may be able to give movies based on actors, directors, description etc.

df['combined_infos'] = [f"{row['type']} {row['title']} {row['director']} {row['cast']} {row['description']} {row['country']}" for _, row in df.iterrows()]
df.head(2)
```

```python
import re
from wordcloud import WordCloud
import nltk
nltk.download(['stopwords',
               'punkt',
               'wordnet',
               'omw-1.4',
               'vader_lexicon'])
from nltk import word_tokenize, sent_tokenize
from nltk.stem import LancasterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import TweetTokenizer
%matplotlib inline
```

```python
tokenizer = nltk.tokenize.RegexpTokenizer('[a-zA-Z0-9\']+')
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

```python
stemmer = nltk.stem.PorterStemmer()
```

```python
def preprocess_text(text):
    tokenized_document = nltk.tokenize.RegexpTokenizer('[a-zA-Z0-9\']+').tokenize(text) # Tokenize
    cleaned_tokens = [word.lower() for word in tokenized_document if word.lower() not in stop_words] # Remove
    stemmed_text = cleaned_tokens #[ nltk.stem.PorterStemmer().stem(word) for word in cleaned_tokens] # Stemming
    return stemmed_text
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Preprocess the text and create a list of words
words = []
for item in df['combined_infos']:
    words.extend(preprocess_text(str(item)))

# Count the frequency of each word
word_counts = Counter(words)

# Create a word cloud from the bag of words
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(word_counts)

# Display the word cloud using matplotlib
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Inserting an new column "Id" into the dataframe 

df.insert(1, "id", list(range(1, df.shape[0]+1)), True)
```

```python
# First apply applies the preprocess_text function to each element of the 'combined_infos' column. 
# The second apply method applies the join method to each element of the resulting series.
# Formatted text looks like below:

df['formatted_combined_info'] = df['combined_infos'].apply(preprocess_text).apply(' '.join)
df['formatted_combined_info'][5]
```

```python
# The Original combined form looks like below:
# We can clearly analyse the difference. 

df['combined_infos'][5]
```

```python
# we need only the formatted/preprocessed text of the combined features
# So, we have created a new dataframe and copied the contents of  "formatted_combined_info" and "Id" columns.

df_formatted = df[['id', 'formatted_combined_info']]
df_formatted.head(2)
```

```python
# Cosine similarity function accepts matrix of token counts
# contents of formatted_combined_info should be converted to matrix of token counts.

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
feature_count_matrix = vectorizer.fit_transform(df_formatted['formatted_combined_info'])
```

```python
feature_count_matrix
```

```python
# Steps to find the cosine similarity matrix

from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(feature_count_matrix)
cosine_similarity_matrix
```

```python
cosine_similarity_matrix.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(25, 20))
# create the heatmap
sns.heatmap(cosine_similarity_matrix, cmap='coolwarm')

# display the heatmap
plt.show()
```
